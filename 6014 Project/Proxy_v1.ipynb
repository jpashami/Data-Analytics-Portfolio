```python
proxy_url = 'https://proxylist.geonode.com/api/proxy-list?limit=50&page=1&sort_by=lastChecked&sort_type=desc'
req= requests.get(proxy_url)
#print(req.json())
json_data = req.json()
proxy_datalist = json_data['data']
for proxy in proxy_datalist:
    proxy_address = proxy['ip']+':'+proxy['port']
    extract(proxy_address)
    #print()

def extract(proxy):
    #this was for when we took a list into the function, without conc futures.
    #proxy = random.choice(proxylist)
    #headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:80.0) Gecko/20100101 Firefox/80.0'}
    try:
        #change the url to https://httpbin.org/ip that doesnt block anything
        r = requests.get('https://httpbin.org/ip', proxies={'http' : proxy,'https': proxy}, timeout=2)
        print(r.json(), ' | Works')
    except:
        print(proxy , 'Faild!')
        pass
    return proxy
```
